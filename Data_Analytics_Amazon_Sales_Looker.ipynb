# Amazon India Sales

##Problem Summary

After a discussion with the sales team, they advised they need a dashboard to best display their orders for the 2nd quarter of 2022.

The dashboard should display:

1.   Total number of orders for the period
2.   Count of orders placed by date
3.   Quantity of orders by clothing type
4.   Order status percentages
5.   An estimate of shipment locations


I proposed to create the following:

1.   A scorecard to display the total number of orders
2.   A linechart to display the orders placed by date
3.   A barchart to display the quantity of orders by clothing type
4.   A barchart to display the order status percentages
5.   A Google maps insert to display the shipment locations

Additionally, I proposed to include a filter where they could adjust the dashboard by date if required.




##Data Analysis & Visualisation

I received the dataset as attached: [dataset](https://www.kaggle.com/datasets/thedevastator/unlock-profits-with-e-commerce-sales-data) and followed the steps:


1.   Loaded the date into Google BigQuery
2.   Cleaned the raw data transforming into enriched data
3.   Analysed the enriched data in Google Looker
4.   Built the dashboard
